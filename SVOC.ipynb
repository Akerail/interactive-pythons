```python
import numpy as np
import time
import scipy as sp
import matplotlib.pyplot as plt
import cmath as cm
%matplotlib inline

ti=time.clock()
```

```python
L = 10**(-9)/1.7
C = 10**(-12)/1.7
R0 = 100

#Hranice
ωmin = 1 #GHz
ωmax = 100
N = 5000
dω = (ωmax - ωmin)/N

#Polia
Z = np.zeros((N,1))
ω = np.zeros((N,1))

plt.figure(figsize=(10,6))

for k in range (10):
    α = 5*k
    for i in range (N):
        ω[i,0] = ωmin + i*dω
        ωreal = ω[i,0]*10**(9)
        R = R0 + α*ω[i,0]
        Z[i,0] = 1/(np.sqrt(1/R**2 + (1/(ωreal*L) - ωreal*C)**2))
    plt.plot(ω, Z, linewidth = 2)

plt.axis([20,90,0,1.2*max(Z)])
plt.xlabel('$ω[GHz]$', fontsize = 15)
plt.ylabel('$Z$', fontsize = 15)

```
