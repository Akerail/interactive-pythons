```python
import numpy as np
import time
import scipy as sp
import matplotlib.pyplot as plt
import cmath as cm
%matplotlib inline
```

```python
#Parameters
kB = 1.38*10**(-23)
m = 10**(-27)
ω = 10**(14)
N = 10
a = 3
A0 = 10**(10)*np.sqrt(3*kB/(m*ω*ω))
```

# Shift of magnitude *A*  in random direction

```python
# iterations, steps
M = 500               
dT = 25                   
NT = 320

T = np.zeros((NT,1))                         
Iteor = np.zeros((NT,1))                     
Isim = np.zeros((NT,1),dtype=np.complex_)


def Teoria (H,K,L):
    for t in range (NT):        
        T[t,0] = t*dT
        G = 2*np.pi*np.sqrt(H*H+K*K+L*L)/a
        A = A0*np.sqrt(T[t,0])
        Iteor[t,0] = np.e**(-A*A*G*G/3)
    return Iteor


def Intenzita (H,K,L):
    dT = 25/H**2                    
    for t in range (NT):
        Amplituda = 0               
        T[t,0] = t*dT
        A = A0*np.sqrt(T[t,0])
        #import pdb
        #pdb.set_trace()
        Suma = 0
        for i in range (N):
            for j in range (N):
                for k in range (N):                                
                    FirstE = np.e**(-1j*2*np.pi*(i*H+j*K+k*L))
                    Suma = Suma + FirstE

        mean = 0                                           
        for m in range (M):
            ϕ = 2*np.pi*np.random.rand(1)
            θ = np.pi*np.random.rand(1)
            Ux = A*np.sin(θ)*np.cos(ϕ)
            Uy = A*np.sin(θ)*np.sin(ϕ)
            Uz = A*np.cos(θ)
            SecondE = np.e**(-1j*2*np.pi*(H*Ux+K*Uy+L*Uz)/a)
            mean = mean + SecondE/M 

        Amplituda = Suma*mean/N**(3)
        Isim[t,0] = (abs(Amplituda))**2
    return (Isim.real)
```

# Plots

```python
ti=time.clock()

for i in range (1,3):
    for j in range (3):
        for k in range (3):
            if i>=j and i>=k and j>=k:
                plt.figure(figsize=(15,6))
                plt.xlabel('$T, ($'+str(i)+str(j)+str(k)+'$)$', fontsize = 15)
                plt.ylabel('$I(T)$', fontsize = 15)
                plt.axis([1, dT*NT/i**2,0,1.1])
        
                plt.plot(T, Intenzita(i,j,0), linewidth = 2)
                plt.plot(T, Teoria(i,j,0), linewidth = 2)


print('NUMPY time', time.clock()-ti, 'seconds')
```
